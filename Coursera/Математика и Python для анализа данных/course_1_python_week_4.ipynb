```python
import numpy as np
from scipy.stats import norm
from matplotlib import pyplot as plt
%matplotlib inline
```

Triangular distribution See: https://en.wikipedia.org/wiki/Triangular_distribution

```python
def build_plot(n, subsets):
    sets = np.random.triangular(0, 1, 2, size = (subsets,n))
    
    means = np.sort(np.sum(sets, axis = 1) / n) 
    
    fit = norm.pdf(means, 1, np.sqrt(1./(6* n))) 
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.plot(means, fit,'-')
    plt.hist(means, bins = 7, normed=True)
    plt.xlim((0.8, 1.5))
```

For Triangular distribution with parameters a = 0, b = 2, c = 1 => mean = 1, variance = 1/6. So mean ~N(1, 1/(6 * n)), n = number of experiments

```python
build_plot(n = 100, subsets = 1000)
```

```python
build_plot(n = 500, subsets = 1000)
```
